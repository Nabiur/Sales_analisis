```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))  # Set figure size

# Get order count
orders_count = df['Country'].value_counts()
sales = df.groupby('Country')['sales'].sum().loc[orders_count.index]

# Define bar width
bar_width = 0.4  

# Create X locations for bars
x = np.arange(len(orders_count))  

# Plot total orders (left bars)
plt.bar(x - bar_width/2, orders_count, width=bar_width, color='red', alpha=0.7, label='Total Orders')

# Plot total sales (right bars)
plt.bar(x + bar_width/2, sales, width=bar_width, color='blue', alpha=0.5, label='Total Sales')

# Customize plot
plt.title('Country-wise Total Orders & Sales')
plt.xlabel('Country')
plt.ylabel('Log Scale Values')
plt.xticks(x, orders_count.index, rotation=45, ha='right')  # Set country labels at x-axis
plt.yscale('log')  # Apply logarithmic scale
plt.legend()
plt.tight_layout()
plt.show()
```

## Task 6: Analyze the sales trends over tim
Analyze the sales trends over time. Identify the busiest months and days of the week in terms of sales.

```python
df.info()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df['month']=df['InvoiceDate'].dt.strftime('%b')
df['day']=df['InvoiceDate'].dt.strftime('%a')
df['num_month']=df['InvoiceDate'].dt.month
df['num_day']=df['InvoiceDate'].dt.dayofweek
df.head()
```

```python
df.info()
```

```python
import calendar
monthwise=df.groupby('num_month')['sales'].sum()
plt.figure(figsize=(12,6))
monthwise.plot(kind='line',marker='o',color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(ticks=range(1,13),labels=calendar.month_name[1:])
plt.title("Total sales per month")

daywise=df.groupby('num_day')['sales'].sum()
plt.figure(figsize=(12,6))
daywise.plot(kind='bar')
plt.xticks(ticks=range(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],rotation=0)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title("Total sales per day")
plt.show()
```

```python
Q1=df.quantile(0.25,numeric_only=True)
Q3=df.quantile(0.75,numeric_only=True)
IQR=Q3-Q1
IQR
```

```python
gf=df.select_dtypes(include='number')
```

```python
low=Q1-IQR*1.5
high=Q3+IQR*1.5
for column in gf.columns:
    print(f'Numbers of outlier in {column}',end=' ')
    print(gf[~(gf[column].between(low[column],high[column]))][column].count())
    
```
